## 1. 파일 시스템 탐색을 위한 리눅스 명령어

- pwd

  + 현재 작업 중인 디렉터리를 보여줌

- is

  + 파일과 디렉터리의 모든 정보를 제공하며 특정 디렉터리와 특정 파일의 내용도 제공

- cd

  + 해당 디렉터리로 이동하는 방법

- mkdir

  + 새 디렉터리를 생성

- rmdir

  + 빈 디렉터리를 삭제할 때 사용

- lsblk

  + 블록 장치의 트리 구조를 나타냄

- mount

  + 기존 파일 시스템으로 마운트 한다 (강력한 터미널 명령)

- df

  + 파일 시스템의 디스크 공간에 대한 필수 정보를 표시하는 강력한 터미널 명령

## 2. 시스템 조작을 위한 리눅스 명령어

- uname

  + 이름, 버전 및 기타 시스템 특정 세부 사항과 같은 시스템 정보를 얻기 위한 기본 명령어

- ps

  + 현재 시스템에서 실행 중인 프로세스를 시각화 할수 있음

- kill

  + 자원 제한으로 인해 멈춘 프로세스를 중지하는 강력한 방법

- service

  + 리눅스 터미널에서 시스템 전체 서비스를 호출

- shutdown

  + halt, init과 함께 시스템을 종료

## 3. 파일관리를 위한 리눅스 명령어

- touch

  + 유효한 빈 파일을 작성
